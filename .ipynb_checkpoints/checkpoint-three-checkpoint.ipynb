## Unnecessary Data

Look for the different types of unnecessary data in your dataset and address it as needed. Make sure to use code comments to illustrate your thought process.

<h1 style="background-color:Blue; color:White; text-align:center">Anime Unnecessary Data Evaluation</h1>

The business issue is focused on helping the staff members in a school district understand their students better through learning and exploring the key elements of anime. Not all of the columns in this data set are relevent for that purpose and the following 12 columns will be dropped:

* broadcast_day

* trailer_url 
* title_english
* title_japanese
* title_synonyms

The Approved column provideds true or false values based on if that record was approved by the currators of the database. The unappoved content has significant errors and missing values. For the purpose of the bussiness issue, records that are unaproved / false will be dropped to so that the data can effectively display the key feature of anime. False values reprsent 14% of the data equating to rough 3,000 rows.

```python
anime_df.drop(['created_at','updated_at','real_start_date','real_end_date','broadcast_day','broadcast_time','main_picture','trailer_url','title_english','title_japanese','title_synonyms'],axis=1, inplace = True)
```

```python
pd.set_option('display.max_columns', None) # This was used to display the columns in the dataframe
anime_df.head()
```

```python
anime_df.drop(anime_df[anime_df['approved'] == False].index, inplace = True)
```

## Missing Data

Test your dataset for missing data and handle it as needed. Make notes in the form of code comments as to your thought process.

<h1 style="background-color:Blue; color:White; text-align:center">Anime Missing Data Evaluation</h1>

```python
number_of_columns = anime_df.shape # I'm viewing the shape to verify the heat map below
number_of_columns
```

```python
# Plot a heatmap of any missing data for Anime Dataset

cols = anime_df.columns[0:27]
colours = ['#000099', '#ffff00'] # specify the colours - yellow is missing. blue is not missing.
sns.heatmap(anime_df[cols].isnull(), cmap=sns.color_palette(colours))
```

Notes About The Data: 

There are a lot of nulls in the 'score' column due to the anime never being scored by viewers. The 'scored_by' column is not null has the value of 0 because no one score it. Since the score column is numerical I'll be changing anyy null value to be 0. 

```python
anime_df1 = anime_df[anime_df.isna().any(axis=1)]
anime_df1
```

```python
anime_df['score'] = anime_df['score'].fillna('0')
```

```python
anime_df['source'] = anime_df['source'].fillna('unknown')
```

```python
anime_df['synopsis'] = anime_df['synopsis'].fillna('Not Available')
```
